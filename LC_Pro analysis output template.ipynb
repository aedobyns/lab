```python
ls = (data_corr[data_corr > 0.95].count())  #use this to conditionally count correlation higher than 0.95
foo = ls[ls>4] # get a Series that have values greater than 4
titles = foo.index #get the row indexes (which are the col headers)

data_high = data[titles] #filter original dataset to only contain cells that meet above criteria
data_high
```

```python
data_high.plot(subplots=True, legend= False)
plt.show()
```
