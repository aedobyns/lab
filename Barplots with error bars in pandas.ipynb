```python
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
AN = pd.read_csv('/Users/abigaildobyns/Downloads/bars/AN.txt', sep='\t', index_col=0)
AH = pd.read_csv('/Users/abigaildobyns/Downloads/bars/AH.txt', sep='\t', index_col=0)
FN = pd.read_csv('/Users/abigaildobyns/Downloads/bars/FN.txt', sep='\t', index_col=0)
FH = pd.read_csv('/Users/abigaildobyns/Downloads/bars/FH.txt', sep='\t', index_col=0)
```

```python
results = [AN, AH, FN, FH]
data = pd.concat(results) #merge all results files into one dataframe
data
```

```python
treatment = data.groupby(by='subject type') #get it all segregated by treatment+group
average = treatment.mean() #get averages based on group
std = treatment.std() #get std based on group
average
```

```python
#to plot only one column with error bars
average['delta'].plot(kind = 'bar', yerr = std['delta'], color = '#C0C0C0')
plt.title('Delta by group')
plt.xlabel('Age and Treatment Group')
plt.ylabel('Delta value selected')
plt.show()
```

```python
#looks like you need to slice dataframes to only include the columns you want to plot if its more than one.
avg_f = average[['false positives','false negatives']]
std_f = std[['false positives','false negatives']]
```

```python
avg_f.plot(kind = 'bar') #yerr gives an error (example below) not sure why.
plt.title('False findings by group')
plt.xlabel('Age and Treatment Group')
plt.ylabel('Count')
plt.show()
```
